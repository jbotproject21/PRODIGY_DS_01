IMPORT ALL LIBRARIES 

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
```

load the cleaned dataset

```python
clean_data = pd.read_csv('cleaned_dataset.csv')
```

```python
clean_data
```

```python
corr_matrix = clean_data.corr(numeric_only=True)
corr_matrix
```

```python
plt.figure(figsize=(13, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

Bar chart: population by country (for a single year)

```python
year = '2024'

df_sorted = clean_data[['Country Name', year]].dropna().sort_values(by=year, ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(df_sorted['Country Name'], df_sorted[year])
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Population')
plt.title(f'Population by Country in {year}')
plt.show()
```

Bar chart: top 10 countries by population

```python
top10 = clean_data[['Country Name', '2024']].dropna().sort_values(by='2024', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top10['Country Name'], top10['2024'])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Most Populous Countries (2024)')
plt.show()
```

Top 10 countries based on average population (20202023)

```python
top10 = clean_data[['Country Name', '2020', '2021', '2022', '2023']].dropna()

top10['avg_population'] = top10[['2020', '2021', '2022', '2023']].mean(axis=1)

top10 = top10.sort_values(by='avg_population', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top10['Country Name'], top10['avg_population'])
plt.xlabel('Country')
plt.ylabel('Average Population')
plt.title('Top 10 Most Populous Countries (Avg 20202023)')
plt.show()
```
