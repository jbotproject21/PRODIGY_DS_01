# READ THE DATASET 


```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

 # reading csv file 

```python
data = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv', skiprows=4)
```

```python
data 
```

dropping unnecessary columns


```python
df = data.drop(
    columns=['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 69']
)
df.columns
```

checking data information

```python
df.info()
```

checking for missing number 

```python
df.isna()
```

```python
df_clean = df.dropna()
df_clean
```

checking for duplicate numbers 


```python
df_clean.duplicated()
df_clean = df_clean.drop_duplicates()
df_clean

```

removing symbols from the dataset

```python
# Remove $, %, commas, spaces, etc.
df = df_clean.replace({r'[\$,%,.,\s]': ' '}, regex=True)
df
```

```python
df
```

saving the cleaned dataset for analysis

```python
df.to_csv('cleaned_dataset.csv', index=False)
```
